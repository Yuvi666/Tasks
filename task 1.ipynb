```python
# Import necessary libraries
import pandas as pd

# Load the dataset with the correct encoding
file_path = 'games_sales.csv'
df = pd.read_csv(file_path, encoding='latin1')

# Display the first few rows to understand the structure
print("Initial Dataset:")
display(df.head())
```

```python
# Create: Insert new records into the dataset
new_data = {
    'Name': 'New Game',
    'Sales': 5.0,
    'Series': 'New Series',
    'Release': 'Jul-24',
    'Genre': 'Action',
    'Developer': 'New Dev',
    'Publisher': 'New Publisher'
}

# Convert new_data to a DataFrame
new_df = pd.DataFrame([new_data])

# Use concat to add the new row
df = pd.concat([df, new_df], ignore_index=True)
print("\nDataset after adding a new record:")
display(df.tail())  
#Displays the last few rows of the DataFrame to show the newly added record.
# Save the updated dataset back to CSV
df.to_csv(file_path, index=False, encoding='latin1')
#Saves the updated DataFrame back to the CSV file.
```

```python
# Read: Retrieve and display specific records from the dataset
# Example: Fetching all records where Sales > 20
filtered_df = df[df['Sales'] > 20]
print("\nRecords where Sales > 20:")
display(filtered_df)
```

```python
# Update: Modify existing records in the dataset
# Example: Update the 'Sales' of 'New Game' to 10.0
df.loc[df['Name'] == 'New Game', 'Sales'] = 10.0
print("\nDataset after updating 'New Game' sales to 10.0:")
display(df[df['Name'] == 'New Game'])

# Save the updated dataset back to CSV
df.to_csv(file_path, index=False, encoding='latin1')
```

```python
# Delete: Remove specific records from the dataset
# Example: Remove the record for 'New Game'
df = df[df['Name'] != 'New Game']
print("\nDataset after deleting 'New Game':")
display(df.tail())

# Save the updated dataset back to CSV
df.to_csv(file_path, index=False, encoding='latin1')
```
